```python
def clean_empty_rows(data):
    """Clean rows with all NaN except price, sqft and area"""
    # remove columns that are not empty
    # use set for a more efficient run
    set_ = set(data.columns)
    for column in ['price', 'sqft', 'area']:
        set_.remove(column)
    cols = list(set_)
    
    # drop empty columns 
    data.dropna(subset=cols, how='all', inplace=True)
    return data
```

## Imputing Missing Values

Since `type`, `area` and `type` have no missing values, we will likely use these features to impute values onto other features, unless there's a strong correlation from other features, then we will use them first!

Let's look at which feature we should focus on first!

### yearbuilt

It looks like we should fill in `yearbuilt` first. My intuition tells me that `type` may be related to it. For an expensive area such as the Bay Area, my suspicion is that there will be newer `Condo` and `Townhouse` types while `Single Family` and `Multi Family` are older on average.

```python
test = data.loc[:, ['type', 'yearbuilt']]
ax = sns.boxplot(x="type", y="yearbuilt", data=test)
```

It seems that my suspicions are correct. Let's go ahead and impute the `yearbuilt` column based on the median `yearbuilt` of a specific `type`. I will not impute the mean in this scenario as there are lots of outliers on both ends.

```python
def impute_na_by_stats(data, column_category, column_na, how='mean', dtype='float'):
    """
    Impute missing values via mean or median
    Input:
        data            - dataframe in use
        column_category - column from which values are grouped by
        column_na       - column whose missing value needs imputing
        how             - values imputed by mean or median
        dtype           - specify whether to keep as float or round to int
    
    Output:
        none            - designated column has missing values imputed
    """
    for col in data[column_category].unique():
        # impute via mean or median
        if how == 'mean':
            stats = data.loc[data[column_category] == col, column_na].mean()
        else:
            stats = data.loc[data[column_category] == col, column_na].median()
        
        # remove floating precision
        if dtype == 'int':
            stats = stats // 1.0
        
        # assign result to dataframe
        data.loc[(data[column_category] == col) & (data[column_na].isnull()), column_na] = stats
```

```python
impute_na_by_stats(data, 'type', column_na='yearbuilt', how='median', dtype='int')
print_missing_data(data)
```

Perfect! We've imputed the values in `yearbuilt`. 

### bed, bath and sqft

Now let's look at the missing values in `bed`. My first hunch is that I should use `sqft` to impute the values of `bed`, since the large the house, the likelihood that the number of `bed` also goes up!

Looking at the correlation matrix though, the coefficient between `sqft` and `bed` is 0.67 while the coefficient between `bed` and `bath` is 0.83. Moreover, `bath` also have around 40.9% missing data. I also note that there might be a relationship between `type` and `bed`.

Hence, I propose imputing missing values this way:

1. Impute values from `type` into `bed` to reduce the missing values in `bed` to 0%;
2. Impute values from `bed` to `sqft` to reduce the missing values in `sqft` into 0%;
3. Impute values from `bed` to `bath` (highest correlation coefficient).

These three steps would ensure that `bed`, `sqft` and `bath` will have 0% missing values!

Before we can use the `sqft` data, however, note the exponential behavior exhibited by `sqft` vis-a-vis `bed` and `bath`. It is therefore best to apply a log transformation before we proceed to create a feature called `sqft_log`.

```python
data.loc[:, 'sqft_log'] = np.log(data['sqft'])
sns.distplot(data['sqft_log'], fit=norm);
```

That looks great! Now, let's use this log transformation to impute our values in the following order:

1. `type` to `bed`;
2. `bed` to `sqft_log`; and
3. `bed` to `bath.

```python
def impute_na_by_linreg(col_impute, col_na, bool_1, inverse=False, bool_inverse=False):
    """
    Impute missing values by linear regression
    
    Inputs:
        data       - the dataframe
        col_impute - the column from which the data will be imputed from
        col_na     - the column into which missing values will be imputed
        bool_1     - a boolean Series of whether the col_na entries are empty
        inverse    - whether the user wishes to impute missing values from 
                     col_na to col_impute
        bool_2     - a boolean Series of whether the col_impute entries are empty
    
    Output:
        none - imputed missing values in data (with/out inverse values)
    
    """
    # calculate linear regresssion stats
    no_null = (~data[col_impute].isnull()) & (~data[col_na].isnull())
    impute  = data.loc[no_null, col_impute].tolist()
    na      = data.loc[no_null, col_na    ].tolist()
    slope, intercept, r_value, p_value, std_err = stats.linregress(impute, na)
    
    # impute na values
    data.loc[bool_1, col_na] = data[col_impute] * slope + intercept
    
    # impute inverse values
    if inverse:
        data.loc[bool_inverse, col_impute] = (data[col_na] - intercept) / slope
```

```python
test = data.loc[:, ['type', 'bed']]
ax = sns.boxplot(x="type", y="bed", data=test)
```

There seem to be lots of outliers in the `bed` column. Let's use the `median` to impute `bed` values.

```python
impute_na_by_stats(data, 'type', column_na='bed', how='median', dtype='int')
print_missing_data(data)
```

Great! There are no more missing values in `bed`. Let's now use `bed` to impute values in `sqft_log` and `bath`.

```python
impute_na_by_linreg(col_impute='bed', col_na='sqft_log', bool_1=data['sqft_log'].isnull())
print_missing_data(data)
```

```python
impute_na_by_linreg(col_impute='bed', col_na='bath', bool_1=data['bath'].isnull())
print_missing_data(data)
```

Before we move on, let's clean up the values in `bed` and `bath`. In housing, typically agent listings will include half bathrooms. Let's check to see if there is a half bathroom listing!

## One-Hot Encoding with NA

Let's one hot encode the categorical features. We're not including `parking2` and `parking3` as we will add them back into the one-hot encoded columns in `parking1`.

```python
categories_null = ['cooling', 'heating', 'parking1']
categories_no_null = ['area', 'type']
data = pd.get_dummies(data, columns=categories_null, dummy_na=True)
data = pd.get_dummies(data, columns=categories_no_null)
data.head()
```

# 4. Data Modeling

Let's separate the target variable `price_log` from the other variables and train the model.

```python
def train_model(X, y):
    """Score the linear regression model with r squared"""
    # train linear regression model
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .30, random_state=42)
    lm_model = LinearRegression(normalize=True)
    lm_model.fit(X_train, y_train)
    
    # predict using your model
    y_train_hat = lm_model.predict(X_train)
    y_test_hat  = lm_model.predict(X_test )

    # score using your model
    train_score = r2_score(y_train, y_train_hat)
    test_score  = r2_score(y_test , y_test_hat )
    
    # print result
    message_train = "The r-squared score for the model was {:.3f} on {} values of train data."
    message_test = "The r-squared score for the model was {:.3f} on {} values of test data."
    print(message_train.format(r2_score(y_train, y_train_hat), len(y_train)))
    print(message_test.format(r2_score(y_test, y_test_hat), len(y_test)))
    
    return lm_model, X_train
```

```python
# set target variable from others
columns = list(data.columns)
columns.remove('price_log')
X = data[columns]
y = data['price_log']
lm_model, X_train = train_model(X, y);
```

Not bad! I would typically be suspicious of models that are too accurate but let's break them down!

# 5. Evaluate Results

Let's look how how the features contributed to the final result!

```python
def coef_weights(coefficients, X_train):
    '''
    INPUT:
    coefficients - the coefficients of the linear model 
    X_train - the training data, so the column names can be used
    
    OUTPUT:
    coefs_df - a dataframe holding the coefficient, estimate, and abs(estimate)
    
    Provides a dataframe that can be used to understand the most influential coefficients
    in a linear model by providing the coefficient estimates along with the name of the 
    variable attached to the coefficient.
    '''
    coefs_df = pd.DataFrame()
    coefs_df['est_int'] = X_train.columns
    coefs_df['coefs'] = lm_model.coef_
    coefs_df['abs_coefs'] = np.abs(lm_model.coef_)
    coefs_df = coefs_df.sort_values('abs_coefs', ascending=False)
    return coefs_df
```

```python
coef_df = coef_weights(lm_model.coef_, X_train)
coef_df.head(15)
```

It seems that there are superbly high coefficients for some of the features in `type`, `area`, `heating` and `cooling`. It seems though that they are polar opposities.

```python
coef_df.tail(11)
```

From the lower end of the range, it seems that `sqft_log` has a large contribution to the model prediction as well.

We saw from the earlier correlation matrix that `bed` and `bath` are highly correlated with `sqft_log`. There may be collinearity issues there.

Let's see what happens if we remove the smaller coefficients as well as `heating` and `cooling`. I have a hunch that dropping them may not reduce the predictive power of the model.

```python
columns = [
    'age',
    'bed', 
    'bath', 
    'cooling_Central',
    'cooling_Other', 
    'cooling_nan', 
    'heating_Forced air', 
    'heating_Other',
    'heating_nan',
    'parking_spaces', 
    'parking1_Attached Garage',
    'parking1_Carport', 
    'parking1_Detached Garage', 
    'parking1_Off street',
    'parking1_On street', 
    'parking1_nan'
]
train_model(X.drop(columns=columns), y);
```

It did not significantly impact the model! We can use this opportunity to reduce model bloat.

```python
columns = [
    'area_East Bay',
    'area_San Francisco', 
    'area_South Bay', 
    'type_Apartment', 
    'type_Condo',
    'type_Miscellaneous', 
    'type_Multi Family', 
    'type_Single Family',
    'type_Townhouse'
]
train_model(X[columns], y);
```
